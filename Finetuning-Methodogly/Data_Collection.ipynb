```python
import pandas as pd

#Lets load the data and see what sections we have

df = pd.read_csv('/all-the-news-2-1.csv')
unique_sections = df['section'].unique().tolist()
print(unique_sections)

```

```python
#I personally went with politics, but you can choose any section you want; We want to train on one section only since we are only using a 125m parameter model and we want to avoid overfitting
df =df[df['section'] == 'Politics']
```

```python
#Ensyring we have enough data
print(len(df))
```

```python

#We only need these following collumns
df = df[['date', 'title', 'publication', 'author', 'article']]

# Ensure all columns are strings
df = df.astype(str)

#Reduce size if needed
df = df.sample(n=min(20000, len(df)), random_state=42)

# Convert the dataframe to a dictionary
data = {
    'date': df['date'].tolist(),
    'title': df['title'].tolist(),
    'publication': df['publication'].tolist(),
    'author': df['author'].tolist(),
    'article': df['article'].tolist()
}
```

```python
import json
from datasets import Dataset
filename = 'data.json'

# Write the dictionary to a file
with open(filename, 'w') as f:
    json.dump(data, f, indent=4)

print(f"Data saved to {filename}")

with open('data.json', 'r') as file:
    data = json.load(file)

# Convert the data to a Hugging Face dataset
'''We do this step to avoid some unexpected errors that might occur when training the model,'''


dataset = Dataset.from_dict(data)
```
